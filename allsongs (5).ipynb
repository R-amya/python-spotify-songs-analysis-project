```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel(r'C:\Users\Ramya\OneDrive\Documents\Onedrive6\OneDrive\Desktop\New folder\spotify2023newsongs.xlsx')

# Strip leading/trailing spaces from column names and select relevant columns
df.columns = df.columns.str.strip()
df = df[['singer', 'language', 'song_name']]

# Split the singer column and explode the list into individual rows
df['singer'] = df['singer'].str.split('|')
df_exploded = df.explode('singer')

# Create a crosstab of singers and languages
song_count = pd.crosstab(df_exploded['singer'], df_exploded['language'])

# Plot the data
ax = song_count.plot(kind='bar', stacked=False, figsize=(12, 6))

# Add values on top of the bars
for container in ax.containers:
    ax.bar_label(container, label_type='edge')  # Automatically place labels at the edge

# Set the labels and title
plt.xlabel('Singer')
plt.ylabel('Number of Songs')
plt.title('Number of Songs Sung by Artists in Different Languages')
plt.xticks(rotation=90)

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#stacked barchart
df = pd.read_excel(r'C:\Users\Ramya\OneDrive\Documents\Onedrive6\OneDrive\Desktop\New folder\spotify2023newsongs.xlsx')

df.columns = df.columns.str.strip()  
df['released_date'] = pd.to_datetime(df['released_date'], errors='coerce')  # Convert to datetime, ignore errors


df['released_year'] = df['released_date'].dt.year

pivot_streams = pd.pivot_table(df, values='Stream', index='released_year', columns='language', aggfunc='sum', fill_value=0)

ax = pivot_streams.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')

plt.xlabel('Released Year')
plt.ylabel('Total Streams')
plt.title('Total Streams of Songs by Language and Released Year')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

df = pd.read_excel(r'C:\Users\Ramya\OneDrive\Documents\Onedrive6\OneDrive\Desktop\New folder\spotify2023newsongs.xlsx')

df.columns = df.columns.str.strip()

popularity_threshold = df['popularity'].median()
popular_songs = df[df['popularity'] >= popularity_threshold].copy()  

popular_songs.loc[:, 'singer'] = popular_songs['singer'].str.split('|')  
popular_songs_expanded = popular_songs.explode('singer')  

song_count = popular_songs_expanded.groupby(['singer', 'language'], as_index=False)['popularity'].sum()

top_singers = song_count.groupby('singer')['popularity'].sum().nlargest(5).index
top_song_count = song_count[song_count['singer'].isin(top_singers)]

labels = top_song_count['singer'] + ' - ' + top_song_count['popularity'].astype(str)

plt.figure(figsize=(18, 8))
squarify.plot(sizes=top_song_count['popularity'], 
              label=labels, 
              color=plt.cm.tab20.colors[:len(top_song_count)], 
              alpha=0.8)

plt.title('Treemap of Total Popularity of Songs by Top Singers in Different Languages')
plt.axis('off')  
plt.show()  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import seaborn as sns


df = pd.read_excel(r'C:\Users\Ramya\OneDrive\Documents\Onedrive6\OneDrive\Desktop\New folder\spotify2023newsongs.xlsx')


df.columns = df.columns.str.strip()


tempo_by_language = df.groupby('language')['tempo'].sum().reset_index()


colors = sns.color_palette("husl", len(tempo_by_language))


plt.figure(figsize=(8, 8))
plt.pie(tempo_by_language['tempo'], 
        labels=tempo_by_language['language'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.tab20.colors) 

plt.title('Sum of Tempo by Language')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_excel(r'C:\Users\Ramya\OneDrive\Documents\Onedrive6\OneDrive\Desktop\New folder\spotify2023newsongs.xlsx')


df.columns = df.columns.str.strip()

kannada_songs = df[df['language'] == 'Kannada']


if kannada_songs.empty:
    print("No Kannada songs found in the dataset.")
else:
    
    kannada_songs.loc[:, 'singer'] = kannada_songs['singer'].str.split('|')
    kannada_songs_exploded = kannada_songs.explode('singer').reset_index(drop=True)
    song_name_column = kannada_songs.columns[0]  
    num_songs = kannada_songs_exploded[song_name_column].nunique() 
    palette = sns.color_palette("hsv", num_songs)  
    color_map = dict(zip(kannada_songs_exploded[song_name_column].unique(), palette))
    plt.figure(figsize=(20, 8))
    sns.scatterplot(x=song_name_column, y='danceability', data=kannada_songs_exploded, 
                     hue=song_name_column, palette=color_map, s=100)
    plt.xlabel('Song Name')
    plt.ylabel('Danceability Score')
    plt.title('Danceability Scores of Kannada Songs')
    plt.tight_layout()
    plt.show()
```
